```python
model_metrics_df = pd.DataFrame(training_results, columns=results_columns)
model_metrics_df
```

```python
plt.figure(figsize=(12, 5))

sns.set(style='darkgrid')

sns.scatterplot(
    x='weights [MB]',
    y='top1_acc',
    data=model_metrics_df,
    s=100,
    alpha=1,
    hue='model',
    palette='nipy_spectral'
).set_title('Top1 Accuracy vs Training Weights [MB]')

plt.savefig('../Model_Eval_01.webp', bbox_inches='tight')
```
